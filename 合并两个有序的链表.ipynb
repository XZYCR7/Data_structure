```python
def MergeByPointer(pHead1, pHead2):
    
    p1 = pHead1
    p2 = pHead2
    MergeHead = ListNode(None)    # 新链表用于连接输出的节点
    p3 = MergeHead
    
    while p1 or p2:
        
        if p1 != None and p2 == None:    # p2已经到末尾，那么将p1剩下的直接连在新链表后
            p3.next = p1
            return MergeHead.next        # 返回新链表的第二个节点
 
        elif p1 == None and p2 != None:
            p3.next = p2
            return MergeHead.next
        
        elif p1.val > p2.val:
            p3.next = p2                 # 将较小值的节点输出，连接到新链表后
            p2, p3 = p2.next, p3.next    # 更新指针位置
        
        elif p1.val <= p2.val:
            p3.next = p1
            p1, p3 = p1.next, p3.next
    
    return None
if __name__  == "__main__":
    a = ListNode(1)
    b = ListNode(2)
    c = ListNode(3)
    d = ListNode(4)
    e = ListNode(8)
    f = ListNode(6)

    a.next = c
    c.next = e

    b.next = d
    d.next = f


    node = MergeByPointer(a,b)
    while node:
        print(node.val)
        node = node.next
```

```python
def QuickSortForListNode(stack):
    # 快速排序算法
    # 定义递归返回条件
    if len(stack)<=1:
        return stack
    
    base = stack[0]
    left = []
    right = []
    for x in stack[1:]:       # 分治思想，将不大于基准值的放置左边，将大于基准值的放置右边
        if x.val > base.val:
            right.append(x)
        else:
            left.append(x)
    
    return QuickSortForListNode(left) + [base] + QuickSortForListNode(right)
 
def Merge(pHead1, pHead2):
    
    # 特殊情况处理
    if pHead1 == None and pHead2 == None:
        return None
    if pHead1 != None and pHead2 == None:
        return pHead1
    if pHead1 == None and pHead2 != None:
        return pHead2
    
    # 首先将两个链表保存到容器中
    stack = []
    cur = pHead1
    while cur:
        stack.append(cur)
        cur = cur.next
    cur = pHead2
    while cur:
        stack.append(cur)
        cur = cur.next
    
    # 使用排序算法进行排序
    # 快速排序
    stack = QuickSortForListNode(stack)
    
    # 重新整理排序后链表的每一个节点的next指向
    stack[-1].next = None
    for i in range(len(stack)-1):
        stack[i].next = stack[i+1] 
    
    # 返回链表表头
    return stack[0]

if __name__  == "__main__":
    a = ListNode(1)
    b = ListNode(2)
    c = ListNode(3)
    d = ListNode(4)
    e = ListNode(8)
    f = ListNode(6)

    a.next = c
    c.next = e

    b.next = d
    d.next = f


    node = MergeByPointer(a,b)
    while node:
        print(node.val)
        node = node.next
```
