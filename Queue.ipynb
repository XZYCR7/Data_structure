```python
class Solution:
    def __init__(self):
        self.acceptStack = []
        self.outputStack = []
    def push(self,node):
        self.acceptStack.append(node)
    def pop(self):
        if self.outputStack == []:
            while self.acceptStack:
                self.outputStack.append(self.acceptStack.pop())
        if self.outputStack != []:
            return self.outputStack.pop()
        else:
            return None
    
if __name__ == "__main__":
    s = Solution()
    node=["1","2","3"]
    s.push(node)
    print(s.pop())
```
