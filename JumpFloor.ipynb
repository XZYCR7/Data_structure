```python
class Solution:
    def JumpFloor(self, number):
        if number < 1:
            return 0
        if number ==1:
            return 1
        if number ==2:
            return 2
        
        ret = 0
        a = 1
        b = 2
        for i in range(3,number+1):
            ret = a+b
            a = b
            b = ret
            
        return ret
    
if __name__ == "__main__":
    s = Solution()
    ans = s.JumpFloor(4)
    print(ans)
```
