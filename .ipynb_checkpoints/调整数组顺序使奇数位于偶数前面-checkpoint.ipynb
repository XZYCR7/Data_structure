```python
class Solution:
    def reOrder(self,array):
        ret = []
        for i in array:
            if i%2==1:
                ret.append(i)
        for i in array:
            if i%2==0:
                ret.append(i)
        return ret

if __name__ == "__main__":
    s = Solution()
    array01 = [1,3,5,2,6,7,8]
    ans = s.reOrder(array01)
    print(ans)
```
