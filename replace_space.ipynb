```python
class Solution():
    def replace_space(self,string):
        aaa=[]
        for i in range(len(string)-1, -1, -1):
            if string[i] ==" ":
                aaa.append("0")
                aaa.append("2")
                aaa.append("%")
            else:
                aaa.append(string[i])

        aaa.reverse()
        return ''.join(aaa)
        #We are happy
        #aaa=["yppah02%era02%eW"]
    
if __name__ == "__main__":
    s = Solution()
    string01 = 'We are happy'
    ans = s.replace_space(string01)
    print(ans)
```
