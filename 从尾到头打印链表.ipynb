```python
# -*- coding:utf-8 -*-
class ListNode:
   def __init__(self, x):
       self.val = x
       self.next = None

class Solution:
    def printChain(self,node):
        while node:
            print(node.val)
            node = node.next
            
    def printListFromTailToHead(self,listNode):
        ret = []
        pTemp = listNode
        while pTemp:
            ret.insert(0,pTemp.val)
            pTemp = pTemp.next
        return ret
    
if __name__ == "__main__":
    s = Solution()
    #链表初始化
    node1 = ListNode(1)
    node2 = ListNode(2)
    node3 = ListNode(3)
    node1.next = node2
    node2.next = node3

    ans1=s.printChain(node1)
    print(ans1)

    ans = s.printListFromTailToHead(node1)
    print(ans)
```
