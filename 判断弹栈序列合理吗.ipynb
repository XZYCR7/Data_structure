```python
class Solution:
    def IsPopOrder(self,pushV, popV):
        if pushV == [] or len(pushV) != len(popV):
            return None
        stack = []
        index = 0
        for item in pushV:
            stack.append(item)
            while stack and stack[-1] == popV[index]:
                stack.pop()
                index += 1
        return True if stack == [] else False
if __name__ =="__main__":
    s = Solution()
    a = [1,2,3,4,5]
    b = [2,4,5,3,1]
    ans = s.IsPopOrder(a,b)
    print(ans)
```
