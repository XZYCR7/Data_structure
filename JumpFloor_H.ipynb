```python
class Solution():
    def JumpFloorII(self,n):
        if n==1:
            return 1
        ret = 1
        a = 1
        for i in range(2,n+1):
            ret = 2 * a
            a = ret
        return ret
    
if __name__ == "__main__":
    s = Solution()
    ans = s.JumpFloorII(5)
    print(ans)
```
